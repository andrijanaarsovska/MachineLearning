```python
from selenium import webdriver
```

```python
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
```

```python
#naslov, link, kreator, broj na views, time
#views da se raspredeleni no treba transformacija oti se stringovi
```

```python
driver = webdriver.Chrome()
```

```python
driver.get("https://www.youtube.com/feed/trending")
```

```python
driver.page_source
```

```python
videos = driver.find_element(By.ID, "contents")
```

```python
videos.text
```

```python
title = driver.find_element(By.CLASS_NAME, "style-scope ytd-video-renderer")
```

```python
title.text
```

```python
videos= driver.find_elements(By.CLASS_NAME,'style-scope ytd-video-renderer')  
```

```python
data = []
#naslov, link, kreator, broj na views, time

for video in videos:
    title = video.find_element(By.CSS_SELECTOR, '#video-title').text
    creator = video.find_element(By.CSS_SELECTOR, "#text > a").text
    num_views = video.find_element(By.CSS_SELECTOR, "#metadata-line > span:nth-child(3)").text
    time = video.find_element(By.CSS_SELECTOR, "#metadata-line > span:nth-child(4)").text
    link = video.find_element(By.CSS_SELECTOR, "#thumbnail").get_attribute("href")
    dict = {
      "Title" : title,
      "Creator" : creator,
      "NumViews" : num_views,
      "Time" : time,
      "Link" : link,
    }
    data.append(dict)
   
```

```python
data
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
import matplotlib.pyplot as plt
```

```python
def convert_views(view_str):
    if 'K' in view_str:
        return float(view_str.replace('K views', '').replace(' ', '')) * 1000
    elif 'M' in view_str:
        return float(view_str.replace('M views', '').replace(' ', '')) * 1_000_000
    else:
        return float(view_str.replace(' views', '').replace(' ', ''))

df['Numeric Views'] = df['NumViews'].apply(convert_views)
plt.hist(df['Numeric Views'] )
plt.show()
```
